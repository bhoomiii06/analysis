```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
n = 500
data = {
    'Date': pd.date_range('2024-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
print(data)
```

```python
df=pd.DataFrame(data)
print(df.head())
print(df.info())
print(df.describe())
print(df['Category'].value_counts())
```

```python

df.dropna(inplace=True)


df.drop_duplicates(inplace=True)

df['Date'] = pd.to_datetime(df['Date'])

df['Likes'] = df['Likes'].astype(int)


print(df.info())
```

```python

sns.histplot(df['Likes'])
plt.show()


sns.boxplot(x='Category', y='Likes', data=df)
plt.show()

mean_likes = df['Likes'].mean()
print("Mean of Likes:", mean_likes)

mean_likes_by_category = df.groupby('Category')['Likes'].mean()
print(mean_likes_by_category)
```

```python
hist_plot = sns.histplot(df['Likes'])
hist_plot.figure.savefig("histogram_likes.png")

box_plot = sns.boxplot(x='Category', y='Likes', data=df)
box_plot.figure.savefig("boxplot_likes.png")
```
